```python
import matplotlib.pyplot as plt

# Data
years = [2012, 2013, 2014, 2015, 2016, 2017, 2018, 2019, 2020, 2021, 2022, 2023]
cpi_rbi = [109.4789969, 110.0178785, 106.6656567, 104.9069734, 104.9482163, 103.3281734,
           103.9388265, 103.7295057, 106.6234368, 105.1314075, 106.6990341, 105.99]
cpi_combined = [100, 112.06316, 97.68098049, 125.3888287, 104.7035686, 122.1741892,
                122.5148174, 107.9819616, 129.5361903, 129.2258255, 133.7183256, 139.3886]

# Plotting
plt.figure(figsize=(10, 6))
plt.plot(years, cpi_rbi, marker='o', label='CPI_RBI')
plt.plot(years, cpi_combined, marker='o', label='CPI_combined')

# Adding labels and title
plt.xlabel('Year')
plt.ylabel('Consumer Price Index (CPI)')
plt.title('Comparison of CPI_RBI and CPI_combined (2012-2023)')
plt.legend()

# Show plot
plt.grid(True)
plt.show()
```
